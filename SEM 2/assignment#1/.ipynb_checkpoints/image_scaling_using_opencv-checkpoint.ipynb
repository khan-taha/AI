```python
import cv2
import numpy as np
import os
from PIL import Image

print('\n')

number_of_images = 20
target_size = (200, 200)

resized_images = np.empty([20,200, 200, 3])

rootDir = 'images'

for dirName, subdirList, fileList in os.walk(rootDir):
    
    print('Found directory: %s' % dirName ,subdirList,fileList)
    
for i,fname in enumerate(fileList):
    
    pil_image = Image.open(rootDir+"\\"+fname)
    print(fname)
    print("prev format ",np.array(pil_image).shape)
    
    resized_images[i] = cv2.resize(np.array(pil_image), target_size)
    
    print("updated format ", resized_images[i].shape)
    print("\n")
```
